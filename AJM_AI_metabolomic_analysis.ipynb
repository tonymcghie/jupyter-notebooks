```python
import pandas as pd
from sklearn.preprocessing import Imputer, MinMaxScaler, LabelEncoder
import numpy as np
```

```python
def load_data():
    data = pd.read_csv('data/TK251_423a_tasq_AJM.csv')
    Y = data['Classification']

    label_encoder = LabelEncoder()
    Y = label_encoder.fit_transform(Y)

    X = data.drop('Sample #', axis=1)
    X = X.drop('Classification', axis=1)

    imputer = Imputer(missing_values=np.nan, strategy='mean', axis=1)
    scaler = MinMaxScaler()
    X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns, index=X.index)
    X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns, index=X.index)

    return X, Y, label_encoder
```
